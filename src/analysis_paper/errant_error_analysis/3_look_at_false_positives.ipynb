```python
import os
import pandas as pd
```

```python
# read error results
df = pd.read_csv('../best_dev_set_error_type_results.csv')
```

```python
df.head()
```

```python
def filter(df, model, split):
    # filter to model == 'gpt-3.5-turbo-0613'
    df = df[df['model'] == model]
    # filter to split == 'wibea-dev'
    df = df[df['split'] == split]
    df = df.sort_values(by=['FP'], ascending=False)

    return df
```
