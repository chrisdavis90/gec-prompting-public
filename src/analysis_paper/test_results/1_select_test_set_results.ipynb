```python
# utility function for inspecting a dataframe
# function to select rows from dataframe based on prompt_type and prompt_index
def select_rows(df, prompt_type=None, prompt_index=None):
    if prompt_type is None:
        return df[df.prompt_index == prompt_index]
    elif prompt_index is None:
        return df[df.prompt_type == prompt_type]
    else:
        return df[(df.prompt_type == prompt_type) & (df.prompt_index == prompt_index)]
```

```python
test2dev_split = {
    'fce-test': 'fce-dev',
    'jfleg-test': 'jfleg-dev',
    'conll14-test': 'wibea-dev',
}
```

```python
from typing import defaultdict
missing_tests = defaultdict(list)
for test_split, dev_split in test2dev_split.items():
    for model in dev_df.model.unique():
        # get prompt_type_index for model in dev_df for dev_split
        prompt_type_index = dev_df[(dev_df.model == model) & (dev_df.split == dev_split)].prompt_type_index.values[0]

        
        # check if there's a row with model and prompt_type_index in test_df for test_split
        if len(test_df[(test_df.model == model) & (test_df.prompt_type_index == prompt_type_index) & (test_df.split == test_split)]) == 0:
            missing_tests[test_split].append((model, prompt_type_index))

        # # check if model is in test_df for test_split
        # if model not in test_df[test_df.split == test_split].model.values:
        #     missing_tests[test_split].append((model))

        # # check if prompt_type_index is in test_df for test_split
        # if prompt_type_index not in test_df[test_df.split == test_split].prompt_type_index.values:
        #     missing_tests[test_split].append((model, prompt_type_index))
```

```python
merged_dfs = []
for test_split, dev_split in test2dev_split.items():
    # join dev_df_fce and test_df_fce on model and prompt_type_index
    df_merged = dev_df[dev_df['split'] == dev_split].merge(test_df[test_df['split'] == test_split], on=['model', 'prompt_type_index'])

    # rename split_x to dev_split and split_y to test_split
    df_merged = df_merged.rename(columns={'split_x': "dev_split", 'split_y': "test_split"})

    # append to merged_dfs
    merged_dfs.append(df_merged)
```

```python
merged_dfs[0]
```

```python
# concatenate all merged_dfs
merged_df = pd.concat(merged_dfs)
```

```python
# count models and splits
merged_df.groupby(['model', 'test_split']).count()
```

```python
# save merged_df to csv
merged_df.to_csv(os.path.join('..', 'test_results.csv'), index=False)
```
