```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read dataframe
df = pd.read_csv(os.path.join('..', 'cefr_results.csv'))
df.head()
```

```python
model_order = ['bloomz-7b1', 'flan-t5-xxl', 'InstructPalmyra-20b',  'opt-iml-max-30b', 'falcon-40b-instruct', 'Llama-2-70b-chat-hf', 'StableBeluga2', 'command', 'gpt-3.5-turbo-0613', 'gpt-4-0613']
cefr_order = ['A', 'B', 'C', 'N']
model_cefr_order = {}
for i, m in enumerate(model_order):
    for j, c in enumerate(cefr_order):
        model_cefr_order[m + '-' + c] = (i * 4 + j) + 1

# create a new column combining model and cefr
df['model_cefr'] = df['model'] + '-' + df['cefr']

# sort the dataframe by model_cefr
# df['model_cefr_order'] = df['model_cefr'].map(model_cefr_order)
df = df.sort_values(by='model_cefr', key=lambda x: x.map(model_cefr_order))
```

```python
df
```

```python
sns.set_theme(style="whitegrid")

# set figure size
plt.figure(figsize=(15, 3))

# rename models
df['model'] = df['model'].replace({
    'flan-t5-xxl': 'Flan-T5-XXL',
    'InstructPalmyra-20b': 'InstructPalmyra-20B',
    'Llama-2-70b-chat-hf': 'Llama-2-70B-chat',
    'bloomz-7b1': 'Bloomz-7B1',
    'command': 'Cohere Command',
    'opt-iml-max-30b': 'OPT-IML-MAX-30B',
    'gpt-3.5-turbo-0613': 'GPT-3.5-turbo',
    'StableBeluga2': 'StableBeluga2',
    'falcon-40b-instruct': 'Falcon-40B-instruct',
    'gpt-4-0613': 'GPT-4'
    })

# plot df
ax = sns.barplot(x="model", y="f05", hue='cefr', data=df)

# rotate x labels
plt.xticks(rotation=90)

# move legend to the top and make horizontal
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.18), ncol=4)

# set y axis label
plt.ylabel("F$_{0.5}$")

# set x axis label
plt.xlabel("Model")

# y-limit
plt.ylim(0.2, 0.6)

# increase font size
plt.tick_params(labelsize=12)

# rotate x labels
plt.xticks(rotation=45, ha='right')

# save figure in high quality
plt.savefig(os.path.join('best_wibea_dev_results_per_cefr.png'), dpi=300, bbox_inches='tight')
```
